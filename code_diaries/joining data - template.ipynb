```python
import pandas as pd
import os
import glob
```

```python
#combine into a csv for Cruickshank lidar
def combine_csv(file_name, directory):

    os.chdir(directory)

    # Match the pattern (‘csv’) and save the list of file names in the ‘all_filenames’ variable.
    extension = 'csv'
    all_filenames = [i for i in glob.glob('*.{}'.format(extension))]

    # combine all files in the list
    combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames])
    combined_csv.drop_duplicates()
    # export to csv
    combined_csv.to_csv(file_name, index=False)

combine_csv("CRU_coreg_summary.csv", r"H:\Cruickshank\2021\troubleshooting_lidar")
```

```python
#Get current working directory, sub in files here
input1 = r"H:\Cruickshank\2021\troubleshooting_lidar\lidar_extract\CRU\all\CRU_all_points.csv"
input2 = r"H:\Cruickshank\2021\troubleshooting_lidar\lidar_extract\CR3\all\Cr3_all_points.csv"
output = r"H:\Cruickshank\2021\troubleshooting_lidar\lidar_extract\compare_coreg_vs_nocoreg.csv"
```

```python
data = pd.read_csv(input1)
data_tojoin = pd.read_csv(input2)
display(data.head())
```

```python
list(data)
data = data[['ACO_Flight_Number','Plot_ID', 'easting','northing','cardinal','distance', 'depth','Lidar']]
display(data)
```

```python
data_tojoin = data_tojoin[['ACO_Flight_Number','Plot_ID','cardinal','distance','Lidar']]
data_tojoin.rename(columns={'Lidar' : "lidar_no_coreg"}, inplace=True)
display(data_tojoin)
```

```python
#set up key for rows to join on
key = ["ACO_Flight_Number", "Plot_ID", "cardinal", "distance"]

data= data.merge(data_tojoin, on=key)
data["difference"]=data["Lidar"] - data["lidar_no_coreg"]
display(data)
```

```python
display(len(data))
data.drop_duplicates(inplace = True)
data = data.round(1)
display(len(data))
display(data)
```

```python
data.to_csv(output, index=False)
```

```python
#generate summary stats
sum = data.groupby("ACO_Flight_Number").agg({
    "depth": ["mean", "std"],
    "Lidar": ["mean", "std"],
    "lidar_no_coreg" : ["mean", "std"],
    "difference" : "mean",
}).round(1)
display(sum)
sum.to_csv(r"H:\Cruickshank\2021\troubleshooting_lidar\lidar_extract\summary_coreg.csv")
```

```python
sum_by_plot = data.groupby(["ACO_Flight_Number", "Plot_ID"]).agg({
    "depth": ["mean", "std"],
    "Lidar": ["mean", "std"],
    "lidar_no_coreg" : ["mean", "std"],
    "difference" : "mean",
}).round(1)
display(sum_by_plot)
sum_by_plot.to_csv(r"H:\Cruickshank\2021\troubleshooting_lidar\lidar_extract\summary_coreg_by_plot.csv")
```
